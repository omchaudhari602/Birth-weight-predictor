```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data_set/babies.csv")
```

```python
y = df['bwt']
```

```python
x = df.iloc[:,1:]
x.head()
```

```python
#train test slpit
## command :<pip install scikit-l

from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train,y_test = tts(x,y, test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
lr = LinearRegression()
lasso = Lasso()
rigde = Ridge()
```

```python
print(x_train.isnull().sum())  # shows NaNs per column
```

```python
print(x_train.isnull().sum())  # shows NaNs per column

```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train)
x_test_imputed = imputer.transform(x_test)

lr = LinearRegression()
lr.fit(x_train_imputed, y_train)
lasso.fit(x_train_imputed,y_train)
rigde.fit(x_train_imputed,y_train)

```

```python
lr_pred = lr.predict(x_test_imputed)
las_pred = lasso.predict(x_test_imputed)
rid_pred = rigde.predict(x_test_imputed)
```

```python
from sklearn.metrics import r2_score , mean_squared_error
```

```python
lr_rscore = r2_score(y_test,lr_pred)
lasso_rscore = r2_score(y_test,las_pred)
rid_rscore = r2_score(y_test,rid_pred)
```

```python
lr_mse =mean_squared_error(y_test,lr_pred)
lasso_mse = mean_squared_error(y_test,las_pred)
rid_mse = mean_squared_error(y_test,rid_pred)
```

```python
print("\nPerformance of linear RegressionModel:")
print("R@score :", lr_rscore)
print("MSE : ", lr_mse)

print("\nPerformance of lasso RegressionModel:")
print("R@score :", lasso_rscore)
print("MSE : ", lasso_mse)

print("\nPerformance of Ridge RegressionModel:")
print("R@score :", rid_rscore)
print("MSE : ", rid_mse)
```
